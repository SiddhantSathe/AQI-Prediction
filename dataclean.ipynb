```python
import aqi

# Convert PM2.5 concentration to IAQI using EPA algorithm
pm25_iaqi = aqi.to_iaqi(aqi.POLLUTANT_PM25, '12', algo=aqi.ALGO_EPA)
print(f"PM2.5 IAQI: {pm25_iaqi}")

# Get overall AQI from multiple pollutant concentrations (default algorithm: EPA)
aqi_values = aqi.to_aqi([
    (aqi.POLLUTANT_PM25, '12'),
    (aqi.POLLUTANT_PM10, '24'),
    (aqi.POLLUTANT_O3_8H, '0.087')
])
print(f"Overall AQI: {aqi_values}")
```
